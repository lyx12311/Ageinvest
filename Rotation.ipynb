```python
ages=importage("LAMOST",dist=True)
dist=importage("ALL",dist=True)
```

```python
gaiaid=readfits('/Users/yl4331/Desktop/FirstYearML/kepler_dr2_1arcsec.fits')
```

```python
prot=pd.read_csv('KeplerRot-LAMOST.csv')
prot=pd.merge(prot,gaiaid[['designation','source_id']],left_on='DR2Name',right_on='designation')
len(prot)
```

```python
print(np.shape(ages))
ages=pd.merge(ages,dist[['X','Y','z','R','age']],left_on='age',right_on='age',how='inner')
print(np.shape(ages))
```

```python
ages_prot=pd.merge(prot,ages,left_on='source_id',right_on='source_id_2',how='inner')
ages_prot
```

```python
plt.plot(prot['Teff_lam'],prot['Prot'],'k.',alpha=0.5,markersize=2)
plt.scatter(ages_prot['Teff_lam'],ages_prot['Prot'],c=ages_prot['age'])
plt.xlim([8000,3000])
```
