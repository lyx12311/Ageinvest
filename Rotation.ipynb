```python
eew_prot=eew_prot.drop_duplicates(subset=['source_id'],keep='first')
len(prot),len(eew_prot)
```

```python
abund=pd.read_hdf('/Users/yl4331/Desktop/CannonAges/LAMOSTabundances/inferred_labels.h5')
```

```python
eew_prot=pd.merge(eew_prot,abund,left_on='designation_y',right_on='designation',how='inner')
np.shape(eew_prot)
```

```python
# get ride of giants
eew_prot=eew_prot.loc[eew_prot['giantmodel']==0].reset_index()
```

```python
plt.figure(figsize=(10,8))

p1=fitpoints([7000,6200],[0,7.5])
p2=fitpoints([6200,6000],[7.5,13.5])

abedge=(p1(eew_prot['Teff_lam'])<eew_prot['Prot'])&(eew_prot['Teff_lam']>6200)
abedge=abedge|((p2(eew_prot['Teff_lam'])<eew_prot['Prot'])&(eew_prot['Teff_lam']<6200))
eew_prot_edge=eew_prot[abedge]
eew_prot_nonedge=eew_prot[~abedge]

plt.scatter(eew_prot_nonedge['Teff_lam'],eew_prot_nonedge['Prot'],c='k',s=2,vmin=4,vmax=5)
plt.scatter(eew_prot_edge['Teff_lam'],eew_prot_edge['Prot'],c='r',s=2,vmin=4,vmax=5)
plt.plot([7000,6200],p1([7000,6200]),'r--',linewidth=3)
plt.plot([6200,5000],p2([6200,5000]),'r--',linewidth=3)

plt.xlim([7500,4000])
plt.ylim([0,50])
```
